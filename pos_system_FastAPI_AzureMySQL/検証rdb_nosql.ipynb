```python
# データ作成に使う関数その１
def generate_uuid(max_length):
    i = max_length
    array = []
    while(i > 0):
        array.append(str(uuid.uuid4()))
        i -=1
    return array


generate_uuid(3)
        
```

```python
# データ作成に使う関数その2

# right_now = datetime.now(ZoneInfo("Asia/Tokyo"))
# now = datetime(right_now.year, right_now.month, right_now.day, 0, 0, 0, 0)

# start_datetime = now - timedelta(days=90)
# oldest = start_datetime.timestamp()

# end_datetime = now - timedelta(seconds=1)
# latest = end_datetime.timestamp()

# start_datetime = now - timedelta(days=30)
# st_30 = start_datetime.timestamp()

# start_datetime = now - timedelta(days=60)
# st_60 = start_datetime.timestamp()

def time_array(day_range,max_length):

    right_now = datetime.now(ZoneInfo("Asia/Tokyo"))
    now = datetime(right_now.year, right_now.month, right_now.day, 0, 0, 0, 0)

    i = max_length
    array = []
    while(i > 0):
        array.append(now - timedelta(days=np.random.choice(day_range)))
        i -= 1
    return array

time_array(2000,10)
```

```python
# データ作成に使う関数その3
# パスワード配列生成
def password_array(max_length):

    i = max_length
    array = []
    while(i > 0):
        n = random.choice(range(5,15))
        random_list = [random.choice(string.ascii_letters + string.digits) for n1 in range(n)]
        random_chr = "".join(random_list)
        array.append(random_chr)
        i -= 1
    return array

password_array(10)
```

```python
# データ作成に使う関数その4
def invitation_code_array(max_length):

    i = max_length
    array = []
    while(i > 0):
        n = random.choice(range(5,15))
        random_list = [random.choice(string.digits) for n1 in range(n)]
        random_chr = "".join(random_list)
        array.append(random_chr)
        i -= 1
    return array

invitation_code_array(10)
```
